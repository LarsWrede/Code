```python
# SVD
# Jan Nagler, Sept 2021
#import sympy as sympy #symbolic math lib
import numpy as np #powerful numerics lib
import matplotlib.pyplot as plt #plot 
from scipy import * #basic computing
from scipy.linalg import svd #linear algebra operations
from PIL import Image #image library "PILLOW"
from scipy import ndimage
from scipy import misc
```

```python
# compress an image using singular value decomposition 
def my_compress (k): #k equals the numebr of singular values we take
    img = misc.face(gray=True)  # load a racoon, predefined
    U, S, V = svd( img ) # based on ordered singular values (as in class)
    img_output = np.dot(U[:,:k], np.dot(np.diag(S[:k]),V[:k,:]))
    print (k)
    #print (100.0-100.0*(size(U[:,:k])+size(S[:k])+size(V[:k,:]))/size(img))
    plt.imshow(img_output, cmap = 'gray')
    plt.show()
```

```python
for k in range (1,50,5):
    my_compress(k)
```
